```python
#Importamos la libreria para controlar la base de datos
import pymysql
from reportlab.pdfgen import canvas
import time
#Importamos la libreria tkinter
import tkinter as tk
#Importamos todo de la libreria tkinter
from tkinter import *
    
#Tratamos de conectar a la base de datos
try:
    db = pymysql.connect("localhost", "root", "", "automatas2")
    print("Conexión establecida")
except:
    print("Conexión no establecida")
        
#Crear la ventana
root = tk.Tk()
root.config(bd=20)

#Declarando las varibles que se van a necesitar
iden = StringVar()
nombre = StringVar()
confirm = StringVar()
    
def insertar():
    cursor = db.cursor()
    insert = 'INSERT INTO keyword VALUES('+str(iden.get())+', "'+str(nombre.get())+'");'
    cursor.execute(insert)
    db.commit()
    confirm.set(str(nombre.get())+" se agregó")
    limpiar()
    
def eliminar():
    try:
        cursor = db.cursor()
        delete = 'DELETE FROM `keyword` WHERE id = '+str(iden.get())+' and nombre = '+str(nombre.get())
        cursor.execute(delete)
        db.commit()
        confirm.set(str(nombre.get())+" se eliminó")
        limpiar()
    except:
        confirm.set("Nombre o ID incorrectos")
        limpiar()
    
def modificar():
    try:
        cursor = db.cursor()
        update = "UPDATE `keyword` SET nombre = '" + nombre.get() + "' WHERE id = " + iden.get() + ";"
        cursor.execute(update)
        db.commit()
        confirm.set(str(nombre.get())+" se modificó")
        limpiar()
    except:
        confirm.set("ID incorrecto")
        limpiar()
        
def reporte():
    ml = 10
    t = "                          "
    j = 740

    c = canvas.Canvas('Reporte.pdf')

    c.drawString(ml, 830, "Empresa: La Empresa más Fregona")
    c.drawString(ml, 810, "Reporte generado el: " + time.strftime("%d/%m/%y"))
    c.drawString(ml, 760, "ID" + t + "Nombre")

    
    cursor = db.cursor()
    buscar = "select * from keyword"
    cursor.execute(buscar)
    result = cursor.fetchall()


    for row in result:
        iden = row[0]
        nombre = row[1]
        c.drawString(ml, j, "    {0}".format(iden,nombre) + t + "   {1}".format(iden,nombre))
        j = j - 15

    c.save()
    
def limpiar():
    iden.set("")
    nombre.set("")
    
def salir():
    root.destroy()

def main():
    Label(root, text="SISTEMA DE CONTROL DE USUARIOS").pack()
    Label(root, text="by Axis").pack()
    
    Label(root, text="").pack()  # Separador 
    
    #Etiqueta para el ID
    Label(root, text="ID: ").pack()
    #Cuadro de entrada para teclear el valor del ID
    Entry(root, justify="center", textvariable=iden).pack()

    Label(root, text="Nombre: ").pack()
    Entry(root, justify="center", textvariable=nombre).pack()
    
    Label(root, text="").pack()  # Separador

    Entry(root, justify="center", textvariable=confirm, state="disabled").pack()
    
    Label(root, text="").pack()  # Separador

    Button(root, text="Insertar", command=insertar).pack(side="left")
    Button(root, text="Eliminar", command=eliminar).pack(side="left")
    Button(root, text="Modificar", command=modificar).pack(side="left")
    Button(root, text="Reporte", command=reporte).pack(side="left")
    Button(root, text="Salir", command=salir).pack(side="left")

    # Finalmente bucle de la aplicación
    root.mainloop()

if __name__ == "__main__":
    main()
```
